Cosine similarity is a measure used to determine how similar two vectors are, particularly in the context of text data or other high-dimensional data representations.

cosine_similarity computes the cosine similarity between the TF-IDF vectors of the text documents.

importing dependencies

```python
import numpy as np
import pandas as pd
import difflib # A module providing classes and functions for comparing sequences.
# convert text documents into numerical vectors based on the frequency of words while considering the importance of each word in the entire data.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data collection and preprocessing

```python
# loading the data from the esv file to apandas dataframe
movies= pd.read_csv('/content/movies (1).csv')
```

```python
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

```python
# replacing the null valuess with null string
for feature in selected_features:
   movies[feature]=movies[feature].fillna('')
```

```python
# combining all the 5 selected features
combined_features = movies[ 'genres']+' '+movies[ 'keywords']+' '+movies[ 'tagline']+' '+movies[ 'cast']+movies['director']
```

```python
print(combined_features)
```

```python
# converting the text data to feature vectors
vectorizer =TfidfVectorizer()
```

```python
# Import the TfidfVectorizer class from the correct module
from sklearn.feature_extraction.text import TfidfVectorizer

# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine similarity

```python
similarity = cosine_similarity(feature_vectors)

print(similarity)
```

```python
print(similarity.shape)
```

Getting the movie name from the user

```python
# getting the movie name from the user
movie_name = input(' Enter your favourite movie name : ')
```

```python
# creating a list with all the movie names given in the dataset
list_of_all_titles = movies['title'].tolist()
print (list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title
index_of_the_movie= movies[movies.title == close_match]['index']. values [0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies
similarity_score = list (enumerate(similarity[index_of_the_movie]))
print (similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies= sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
print('Movies suggested for you : \n') #Removed extra space before print statement
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies[movies.iindex==index]['title']. values[0]
    if (i<30):
        print(i, '',title_from_index)
        i+=1
```
